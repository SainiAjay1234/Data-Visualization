```python
import numpy as np
x=np.arange(1,11)
y=np.random.randint(100,size=10)
z=np.random.randint(100,size=10)
x,y,z
```

```python
x1=np.arange(1,11)
y1=np.random.randint(100,size=10)
z1=np.random.randint(100,size=10)
x1,y1,z1
```

```python
fig=plt.figure(figsize=(10,10))
ax=fig.add_subplot(111,projection="3d") #111 is 1=row,1=column,1=plot number
ax.scatter(x,y,z,color="r",marker="D",s=10,label="sample")
ax.set_xlabel("x-axis")
ax.set_ylabel("y-axis")
ax.set_zlabel("z-axis")
plt.legend()
plt.show()
```

```python
fig=plt.figure(figsize=(10,10))
ax1=fig.add_subplot(121,projection="3d") 
ax1.scatter(x,y,z,color="r",marker="D",s=10,label="sample")
ax1.set_xlabel("x-axis")
ax1.set_ylabel("y-axis")
ax1.set_zlabel("z-axis")
ax2=fig.add_subplot(122,projection="3d") 
ax2.scatter(x,y,z,color="r",marker="D",s=30,label="sample")
ax2.scatter(x1,y1,z1,color="g",marker="D",s=30,label="sample")
ax2.set_xlabel("x-axis")
ax2.set_ylabel("y-axis")
ax2.set_zlabel("z-axis")
plt.legend()
plt.show()
```
